```python
import sys
from pathlib import Path

# Добавляем путь к библиотеке
notebook_dir = Path().absolute()
project_root = notebook_dir.parent
sys.path.insert(0, str(project_root / 'lab3'))

from src.economic_effect import (
    calculate_cash_flows,
    calculate_total_effect,
    calculate_npv,
    analyze_project,
    InvalidInputError,
    NegativeValueError
)

def test_calculations():
    """Тестирование основных расчётов."""
    
    # Тест 1: постоянные потоки
    cf_constant = calculate_cash_flows(100, 80, periods=3)
    assert cf_constant == [20.0, 20.0, 20.0], "Ошибка в постоянных потоках"
    
    # Тест 2: переменные потоки
    cf_variable = calculate_cash_flows([100, 120, 150], [80, 90, 100], periods=3)
    assert cf_variable == [20.0, 30.0, 50.0], "Ошибка в переменных потоках"
    
    # Тест 3: общий эффект
    total = calculate_total_effect([20, 30, 50])
    assert total == 100.0, "Ошибка в расчёте общего эффекта"
    
    # Тест 4: NPV
    npv = calculate_npv([20, 30, 50], 0.1, initial_investment=100)
    expected_npv = -19.46
    assert abs(npv - expected_npv) < 0.1, f"Ошибка в NPV: {npv} != {expected_npv}"
    
    # Тест 5: полный анализ
    result = analyze_project(
        benefits=[100, 120, 150],
        costs=[80, 90, 100],
        discount_rate=0.1,
        initial_investment=100,
        periods=3
    )
    assert result['npv'] == npv, "Ошибка в анализе проекта"
    
    print(" Все базовые тесты пройдены")

def test_error_handling():
    """Тестирование обработки ошибок."""
    
    # Тест на некорректный тип данных
    try:
        calculate_cash_flows([100, "двести"], 80, periods=2)
        raise AssertionError("Должна была возникнуть ошибка InvalidInputError")
    except InvalidInputError as e:
        print(f" Корректно поймана ошибка: {e}")
    
    # Тест на отрицательные значения
    try:
        calculate_cash_flows(100, -50, periods=1)
        raise AssertionError("Должна была возникнуть ошибка NegativeValueError")
    except NegativeValueError as e:
        print(f" Корректно поймана ошибка: {e}")

if __name__ == "__main__":
    print("Тестирование библиотеки economic_effect")
    print("-" * 40)
    
    test_calculations()
    test_error_handling()
    
    print("-" * 40)
    print("Все тесты завершены")
```
